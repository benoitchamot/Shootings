# Get information about criminal in each state
- Source: https://www.paperprisons.org/statistics.html
- Colleen V. Chien, America's Paper Prisons: The Second Chance Gap, 119 Mich. L. Rev. 519 (2020) (Appendix)
- Note: these are only **samples**

```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Open automated browser
browser = Browser('chrome')
url = 'https://www.paperprisons.org/statistics.html'
browser.visit(url)
```

```python
# Create a Beautiful Soup object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# End the automated browsing session
browser.quit()
```

```python
tables = soup.find_all('table')
table = tables[0]

dataframes = []

for table in tables:
    dataframes.append(pd.read_html(str(table))[0])
```

## Clean DataFrames
- Remove the rows that contains the total for each column or other aggregation data

```python
dataframes[0] = dataframes[0].drop(dataframes[0][dataframes[0]['State'] == 'Total'].index)
dataframes[0].head(3)
```

```python
dataframes[1] = dataframes[1].drop(dataframes[1][dataframes[1]['State'] == 'Total'].index)
dataframes[1] = dataframes[1].drop(dataframes[1][dataframes[1]['State'] == 'Share of Total Count'].index)
dataframes[1].head(3)
```

```python
dataframes[2] = dataframes[2].drop(dataframes[2][dataframes[2]['State'] == 'Total'].index)
dataframes[2].head(3)
```

```python
dataframes[3] = dataframes[3].drop(dataframes[3][dataframes[3]['State'] == 'Total'].index)
dataframes[3].head(3)
```
